```python
def get_Track_ICO_API(df):
    '''
    Returns a DF of information on ICOs from TrackICO API
    
    '''
    
    ICO_df = pd.DataFrame()
    
    for ticker in list(df.ticker):

        response = requests.get(f"https://api.trackico.io/v1/icos/all/?page=1&page_size=1&q={ticker}")

        if response.json()['count'] == 0:
            continue

        else:
            results = pd.DataFrame((response.json()['results']))
            ICO_df = pd.concat([results,ICO_df])
    
    ICO_df.drop(['slug','id','type','status','profile_url','logo_url','rating'], axis=1, inplace=True)
    #reordered_columns = 
    #ICO_df.reindex(columns=reordered_columns)
#     ICO_df['ico_start'] = ICO_df['ico_start'].replace({'None': 0})
#     ICO_df['ico_end'] = ICO_df['ico_end'].replace({'None': 0})
#     ICO_df['Start'] = ICO_df['ico_start'].apply(lambda x: dt.strptime(x, "%Y-%m-%d %H:%M:%S"))
#     ICO_df['End'] = ICO_df['ico_end'].apply(lambda x: dt.strptime(x, "%Y-%m-%d %H:%M:%S"))
    #Create a duration column from the start and end dates
#    ICO_df["Duration"] = ICO_df['End'] - ICO_df['Start']
    ICO_df.reset_index(drop=True, inplace = True)
    ICO_df.set_index('ticker', inplace = True)
    
    return ICO_df
            #df = pd.DataFrame(response.json()['results'][0].values())
    
get_Track_ICO_API(df)
```
